```python
#Red an image and convert it to Binary image
import cv2
img=cv2.imread('new3.jpg',2)
ret, bw_img=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
bw=cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
cv2.imshow("Binary", bw_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#write a program to generate mirror image flip horizonatlly and flip vertically using loop in python

import cv2
import numpy as np
from matplotlib import pyplot as plt
img_orginal=cv2.imread('image5.jpg')
img=cv2.cvtColor(img_orginal, cv2.COLOR_BGR2RGB)
x,y,z=np.shape(img)
mirror_img=np.zeros((x,y,z),dtype=int)
for i in range(0,x):
    k=0
    for j in range(y-1,0,-1):
        mirror_img[i][k][0]=img[i][j][0]
        mirror_img[i][k][1]=img[i][j][1]
        mirror_img[i][k][2]=img[i][j][2]
        k=k+1
        
    cv2.imwrite('image2.jpeg',mirror_img)
    plt.subplot(1,2,1)
    plt.title('Orginal Image')
    plt.imshow(img)
    plt.subplot(1,2,2)
    plt.title('Mirror_img')
    plt.imshow(mirror_img)
    
    
```

```python
#write a python program to find the complement of an image
import cv2
from matplotlib import pyplot as plt
img=cv2.imread('image1.jpeg')
comp_image=255-img
plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Complementary image')
plt.imshow(comp_image)
```

```python
#write a python program to rotate an image
from PIL import Image
from matplotlib import pyplot as plt
Orginal_image=Image.open('image5.jpg')
rotated_image1=Orginal_image.rotate(60)
plt.subplot(1,2,1)
plt.title('Orginal Image')
plt.imshow(Orginal_image)
plt.subplot(1,2,2)
plt.title('Roated image')
plt.imshow(rotated_image1)
```

```python
#write a python code to show the Red Green Blue components of an RGB image separetly
import cv2
import numpy as np
from matplotlib import pyplot as plt
rgb_img = cv2.imread('image5.jpg')
x,y,z = np.shape(rgb_img)
red = np.zeros((x,y,z),dtype=int)
green = np.zeros((x,y,z),dtype=int)
blue = np.zeros((x,y,z),dtype=int)
for i in range(0,x):
    for j in range(0,y):
        red[i][j][0] = rgb_img[i][j][0]
        green[i][j][1]= rgb_img[i][j][1]
        blue[i][j][2] = rgb_img[i][j][2]
img= cv2.cvtColor(rgb_img, cv2.COLOR_RGB2BGR)
plt.subplot(2,2,1)
plt.title('original image')
plt.imshow(img)
plt.axis('off')
plt.subplot(2,2,2)
plt.title('Red')
plt.imshow(red)
plt.axis('off')
plt.subplot(2,2,3)
plt.title('Green')
plt.imshow(green)
plt.axis('off')
plt.subplot(2,2,4)
plt.title('Blue')
plt.imshow(blue)
plt.axis('off')
```
